```python
from numpy import mean, absolute 
  
#data = [75, 69, 56, 46, 47, 79, 92, 97, 89, 88, 36, 96, 105, 32, 116, 101, 79, 93, 91, 112] 
  
# Assume any point A about which  
# absolute deviation is to be calculated 
A = 6
  
sum = 0  # Initialize sum to 0 
  
# Absolute deviation calculation 
  
for i in range(len(df['fixed acidity'])): 
    av = absolute(df[i] - A)   # Absolute value of the differences  
                                 # of each data point and A 
  
    # Summing all those absolute values 
    sum = sum + av                
  
# Sum divided by length of data yields 
# the absolute deviation 
print(sum / len(df))              
```

## 5. Covariance and Correlation

1. Covariance is a numerical measure that indicates the inter-dependency between two variables. 
2. Covariance indicates how two variables behave together in relation to their averages. 
3. A covariance of 0 indicates that the variables are totally independant, while a high and positive covariance value means that a variable is big when the other is big. Analogously, a negative covariance with a high absolute value means that one variable is big when the other is small.
4. Covariance can be calculated through the cov() function. 

### 5.1 Check the Covariance and Correlation scores . write your Observations?

#### Covariance

```python
df.cov()
```

```python
np.cov(df['fixed acidity'], df['volatile acidity'])
```

#### Observation:
  

By Interpreting covariance of the 'fixed acidity' and 'volatile acidity' it is hard to gain any meaning of the covariance betweeen two variables. to support this correlation between two variables is introduced.
